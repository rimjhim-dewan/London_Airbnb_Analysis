```python
df['inexpensive']=0
df['mid']=0
df['high']=0
df['next_level']=0
```

```python
for i in range(len(df)):
    curr_words=df['property_type'][i].split()
    
    if('shared' in curr_words):
        df['inexpensive'][i]=1
    elif('private' in curr_words):
        df['mid'][i]=1
    elif('entire' in curr_words):
        df['high'][i]=1
    elif('condo' in curr_words or 'villa' in curr_words):
        df['next_level'][i]=1
    else:
        df['mid'][i]=1
```
